```python
import pandas as pd
from sklearn.linear_model import  LinearRegression
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('car driving risk analysis.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
plt.scatter(df['speed'],df['risk'])
plt.xlabel('Speed')
plt.ylabel('Risk')
plt.title('Car Driving Risk')
```

```python
x = df[['speed']]
```

```python
Reg = LinearRegression()
```

```python
Reg.fit(x,df.risk)
```

```python
Reg.predict([[100]])
```

```python
Reg.coef_
```
